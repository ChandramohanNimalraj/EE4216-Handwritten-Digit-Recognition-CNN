# CNN Model Experiments and Comparison

In this section, five different CNN model configurations are developed and evaluated using the MNIST dataset. The purpose of these experiments is to compare how changes in the CNN architecture and training parameters affect model performance.

The assignment requires experiments by changing CNN-related parameters such as convolution layers, kernel sizes, optimizers, activation functions, and model complexity.

The five models used in this notebook are:

| Model | Description | Main Change |
|------|-------------|-------------|
| Model 1 | Baseline CNN Model | 2 Conv layers, 3×3 kernel, Adam, ReLU |
| Model 2 | Kernel Size Comparison | Kernel size changed from 3×3 to 5×5 |
| Model 3 | Convolution Layer Comparison | Added one extra convolution layer |
| Model 4 | Optimizer Comparison | Optimizer changed from Adam to SGD |
| Model 5 | Activation Function Comparison | Activation changed from ReLU to Tanh |

Each model will be trained using the same MNIST training dataset and evaluated using the MNIST test dataset. The results will be compared using test accuracy, test loss, training behavior, and computational complexity.

At the end of these experiments, a comparison table will be created to identify the best-performing CNN model.

## Model 1 – Baseline CNN Model

```python
# ============================================================
# Model 1 – Baseline CNN Model
# ============================================================

# Import required libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# ------------------------------------------------------------
# Load MNIST Dataset
# ------------------------------------------------------------
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Training Images Shape:", x_train.shape)
print("Training Labels Shape:", y_train.shape)
print("Testing Images Shape :", x_test.shape)
print("Testing Labels Shape :", y_test.shape)

# ------------------------------------------------------------
# Display Sample Images
# ------------------------------------------------------------
plt.figure(figsize=(10, 4))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title("Label: " + str(y_train[i]))
    plt.axis('off')

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# Normalize Images
# ------------------------------------------------------------
x_train = x_train / 255.0
x_test = x_test / 255.0

print("Minimum Pixel Value:", x_train.min())
print("Maximum Pixel Value:", x_train.max())

# ------------------------------------------------------------
# Reshape Images for CNN Input
# ------------------------------------------------------------
x_train = x_train.reshape(60000, 28, 28, 1)
x_test = x_test.reshape(10000, 28, 28, 1)

print("Training Images Shape After Reshape:", x_train.shape)
print("Testing Images Shape After Reshape :", x_test.shape)

# ------------------------------------------------------------
# Build Baseline CNN Model
# ------------------------------------------------------------
model_1 = Sequential()

model_1.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model_1.add(MaxPooling2D(pool_size=(2, 2)))

model_1.add(Conv2D(64, (3, 3), activation='relu'))
model_1.add(MaxPooling2D(pool_size=(2, 2)))

model_1.add(Flatten())
model_1.add(Dense(128, activation='relu'))
model_1.add(Dense(10, activation='softmax'))

model_1.summary()

# ------------------------------------------------------------
# Compile Model
# ------------------------------------------------------------
model_1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Model 1 compiled successfully!")

# ------------------------------------------------------------
# Train Model
# ------------------------------------------------------------
start_time = time.time()

history_1 = model_1.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2
)

end_time = time.time()
training_time_1 = end_time - start_time

print("Model 1 Training Time:", training_time_1, "seconds")

# ------------------------------------------------------------
# Plot Accuracy Graph
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(history_1.history['accuracy'], label='Training Accuracy')
plt.plot(history_1.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model 1 - Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# ------------------------------------------------------------
# Plot Loss Graph
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(history_1.history['loss'], label='Training Loss')
plt.plot(history_1.history['val_loss'], label='Validation Loss')
plt.title('Model 1 - Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# ------------------------------------------------------------
# Evaluate Model
# ------------------------------------------------------------
test_loss_1, test_accuracy_1 = model_1.evaluate(x_test, y_test)

print("Model 1 Test Loss:", test_loss_1)
print("Model 1 Test Accuracy:", test_accuracy_1)

# ------------------------------------------------------------
# Generate Predictions
# ------------------------------------------------------------
predictions_1 = model_1.predict(x_test)
predicted_labels_1 = np.argmax(predictions_1, axis=1)

print("Model 1 predictions generated successfully!")

# ------------------------------------------------------------
# Display Sample Predictions
# ------------------------------------------------------------
plt.figure(figsize=(12, 6))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title(f"Actual: {y_test[i]}\nPred: {predicted_labels_1[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# Confusion Matrix
# ------------------------------------------------------------
cm_1 = confusion_matrix(y_test, predicted_labels_1)

disp_1 = ConfusionMatrixDisplay(confusion_matrix=cm_1)

plt.figure(figsize=(8, 8))
disp_1.plot(cmap='Blues')
plt.title("Model 1 - Confusion Matrix")
plt.show()

# ------------------------------------------------------------
# Misclassified Images
# ------------------------------------------------------------
misclassified_1 = np.where(predicted_labels_1 != y_test)[0]

print("Model 1 Total Misclassified Images:", len(misclassified_1))

plt.figure(figsize=(12, 6))

for i in range(10):
    index = misclassified_1[i]
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[index].reshape(28, 28), cmap='gray')
    plt.title(f"A:{y_test[index]} P:{predicted_labels_1[index]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

# Store misclassified count
misclassified_count_1 = len(misclassified_1)

# ------------------------------------------------------------
# Store Model 1 Results
# ------------------------------------------------------------
model_1_results = {
    "Model": "Model 1 - Baseline",
    "Conv Layers": 2,
    "Kernel Size": "3x3",
    "Optimizer": "Adam",
    "Activation": "ReLU",
    "Dense Neurons": 128,
    "Test Loss": test_loss_1,
    "Test Accuracy": test_accuracy_1,
    "Training Time": training_time_1,
    "Parameters": model_1.count_params(),
    "Misclassified": misclassified_count_1
}

model_1_results


# ------------------------------------------------------------
# Custom Handwritten Testing
# ------------------------------------------------------------

import cv2
import os

image_folder = "handwritten"

image_files = sorted(os.listdir(image_folder))

print("Custom handwritten image files:")
print(image_files)

custom_images = []
custom_predictions = []

for file_name in image_files:
    image_path = os.path.join(image_folder, file_name)

    # Read image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Invert image: black digit on white background -> white digit on black background
    img = 255 - img

    # Threshold image
    _, img = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)

    # Find digit boundary
    coords = cv2.findNonZero(img)
    x, y, w, h = cv2.boundingRect(coords)

    # Crop digit
    img = img[y:y+h, x:x+w]

    # Resize digit to 20x20
    img = cv2.resize(img, (20, 20))

    # Add padding to make 28x28
    img = np.pad(img, ((4, 4), (4, 4)), mode='constant', constant_values=0)

    # Normalize
    img = img / 255.0

    # Reshape for CNN
    img_input = img.reshape(1, 28, 28, 1)

    # Predict using Model 1
    prediction = model_1.predict(img_input)
    predicted_digit = np.argmax(prediction)

    custom_images.append(img)
    custom_predictions.append(predicted_digit)

    print(file_name, "-> Predicted Digit:", predicted_digit)

# Display custom handwritten prediction results
plt.figure(figsize=(10, 4))

for i in range(len(custom_images)):
    plt.subplot(1, len(custom_images), i + 1)
    plt.imshow(custom_images[i], cmap='gray')
    plt.title(f"Pred: {custom_predictions[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Engineering Discussion

The experimental results demonstrate that architectural modifications significantly influence CNN performance.

Model 1 served as the baseline CNN architecture and produced high classification accuracy on the MNIST dataset.

Model 2 increased the convolution kernel size from 3×3 to 5×5. This improved feature extraction slightly but also increased computational complexity.

Model 3 introduced an additional convolution layer. Although it achieved competitive accuracy, the training time and number of parameters increased because of the deeper architecture.

Model 4 replaced the Adam optimizer with SGD. While SGD required less computation, it converged more slowly and achieved lower classification accuracy than Adam.

Model 5 replaced the ReLU activation function with Tanh. The model performed reasonably well but did not outperform the ReLU-based architectures because Tanh is more susceptible to gradient saturation.

Overall, the comparison demonstrates that the choice of optimizer, activation function, kernel size, and network depth all affect CNN performance. Among the evaluated models, the architecture with the best balance between classification accuracy and computational efficiency can be considered the most suitable for handwritten digit recognition.
